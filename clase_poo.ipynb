POO---> tener en cuenta 4 pilares
  1.Encapsulamiento-->proteger los datos
  2.Herencia--------->reutiliza codigo
  3.Polimorfismo----->mismo metodo, diferente comportamiento
  4.Abstraccion------>oculta complejidad

```python
class CuentaBancaria:
  def __init__(self,titular,saldo_inicial):
    self.titular=titular
    self.__saldo=saldo_inicial
  def get_saldo(self):
    return self.__saldo
  def depositar(self,cantidad):
    if cantidad>0:
      self.__saldo+=cantidad
      return f"deposito exitoso. nuevo saldo: ${self.__saldo}"
    return "cantidad invalida"
  def retirar(self,cantidad):
    if 0<cantidad<=self.__saldo:
      self.__saldo-=cantidad
      return f"retiro exitoso nuevo saldo: ${self.__saldo}"
    return "fondos insuficientes o cantidad invalida"
cuenta = CuentaBancaria("Ana",1000)
print(cuenta.get_saldo())
print(cuenta.depositar(500))
print(cuenta.retirar(200))
```

```python
class estudiante:
  def __init__(self,nombre,edad,calificaciones):
    self.__nombre=nombre
    self.__edad=edad
    self.__calificaciones=calificaciones
  def agregar_calificacion(self, calificacion):
    if 0 <= calificacion <= 100:
      self.__calificaciones.append(calificacion)
      return f"Calificación {calificacion} agregada exitosamente."
    else:
      return "Calificación inválida. Debe estar entre 0 y 100."
  def calcular_promedio(self):
    if not self.__calificaciones:
      return "No hay calificaciones para calcular el promedio."
    return sum(self.__calificaciones) / len(self.__calificaciones)
nota=estudiante("prueba",20,[])
print(nota.agregar_calificacion(40))
print(nota.agregar_calificacion(90))
print(nota.agregar_calificacion(40))
print(nota.agregar_calificacion(100))
print(nota.agregar_calificacion(100))
print(nota.calcular_promedio())
```

```python
#crear clase llamada carro, agregar __init__ metodo con una marca parametro y almacenarlo como una propiedad, agregar un metodo llamado mostrar que imprima la marca, crear un objeto c1 de la carro clase con la marca ford, llama al mostrar metodo en c1
class carro:
  def __init__(self,marca):
    self.__marca=marca
  def mostrar(self):
    print({self.__marca})
c1 = carro("Ford")
c1.mostrar()
```

```python
#crear clase estudiante con un __init__ que tome nombre y grado y los almacene como propiedades, crear objeto s1 con el nombre "anna" y la calificacion "A", imprime la calificacion de s1, cambiar la calificacion de s1 a "B", imprimir la calificacion actualizada
class estudiante:
  def __init__(self, nombre, grado):
    self.__nombre = nombre
    self.__grado = grado
  def get_grado(self):
    return self.__grado
  def set_grado(self, nuevo_grado):
    self.__grado = nuevo_grado
s1 = estudiante("Anna", "A")
print(f"Calificación inicial de {s1._estudiante__nombre}: {s1.get_grado()}")
s1.set_grado("B")
print(f"Calificación actualizada de {s1._estudiante__nombre}: {s1.get_grado()}")
```
